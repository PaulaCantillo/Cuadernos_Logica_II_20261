```python
"""
euler_sat.py
============
Generador Lógico de Senderos Eulerianos
Contiene:
  1. Toda la librería de Lógica (Logica_5.py) sin modificaciones funcionales,
     salvo correcciones de bugs menores documentadas inline.
  2. Clase Euler con las 7 reglas en formato clausal optimizado
     (reglas_corregidas.py integradas).
  3. Demostración de WalkSAT sobre R1, R2 y R3 de forma aislada.

Autores originales: Lógica para Ciencias de la Computación — U. Rosario
Integración y correcciones: equipo del proyecto, 2026
"""


# 0. IMPORTS


from itertools import product, combinations
from types   import MethodType
from copy    import deepcopy
from random  import choice, uniform, randint

import numpy as np



# 1. LIBRERÍA DE LÓGICA (Logica_5.py)


# 1.1  Clases de fórmulas


class Formula:

    def __init__(self):
        pass

    def __str__(self):
        if type(self) == Letra:
            return self.letra
        elif type(self) == Negacion:
            return '-' + str(self.subf)
        elif type(self) == Binario:
            return "(" + str(self.left) + self.conectivo + str(self.right) + ")"

    def letras(self):
        if type(self) == Letra:
            return set(self.letra)
        elif type(self) == Negacion:
            return self.subf.letras()
        elif type(self) == Binario:
            return self.left.letras().union(self.right.letras())

    def subforms(self):
        if type(self) == Letra:
            return [str(self)]
        elif type(self) == Negacion:
            return list(set([str(self)] + self.subf.subforms()))
        elif type(self) == Binario:
            return list(set([str(self)] + self.left.subforms() + self.right.subforms()))

    def num_conec(self):
        if type(self) == Letra:
            return 0
        elif type(self) == Negacion:
            return 1 + self.subf.num_conec()
        elif type(self) == Binario:
            return 1 + self.left.num_conec() + self.right.num_conec()

    def valor(self, I):
        if type(self) == Letra:
            return I[self.letra]
        elif type(self) == Negacion:
            return not self.subf.valor(I)
        elif type(self) == Binario:
            if self.conectivo == 'Y':
                return self.left.valor(I) and self.right.valor(I)
            if self.conectivo == 'O':
                return self.left.valor(I) or self.right.valor(I)
            if self.conectivo == '>':
                return not self.left.valor(I) or self.right.valor(I)
            if self.conectivo == '=':
                return (self.left.valor(I) and self.right.valor(I)) or \
                       (not self.left.valor(I) and not self.right.valor(I))

    def SATtabla(self):
        letras = list(self.letras())
        n = len(letras)
        valores = list(product([True, False], repeat=n))
        for v in valores:
            I = {letras[x]: v[x] for x in range(n)}
            if self.valor(I):
                return 'Satisfacible', I
        return 'Insatisfacible', None

    def clasifica_para_tableaux(self):
        if type(self) == Letra:
            return None, 'literal'
        elif type(self) == Negacion:
            if type(self.subf) == Letra:
                return None, 'literal'
            elif type(self.subf) == Negacion:
                return 1, 'alfa'
            elif type(self.subf) == Binario:
                if self.subf.conectivo == 'O':
                    return 3, 'alfa'
                elif self.subf.conectivo == '>':
                    return 4, 'alfa'
                elif self.subf.conectivo == 'Y':
                    return 1, 'beta'
        elif type(self) == Binario:
            if self.conectivo == 'Y':
                return 2, 'alfa'
            elif self.conectivo == 'O':
                return 2, 'beta'
            elif self.conectivo == '>':
                return 3, 'beta'

    def SATtableaux(self):
        estado = nodos_tableaux([self])
        res = estado.es_hoja()
        if res == 'cerrada':
            return 'Insatisfacible', None
        elif res == 'abierta':
            return 'Satisfacible', estado.interp()
        frontera = [estado]
        while len(frontera) > 0:
            estado = frontera.pop(0)
            hijos = estado.expandir()
            for a in hijos:
                if a is not None:
                    res = a.es_hoja()
                    if res == 'abierta':
                        return 'Satisfacible', a.interp()
                    elif res is None:
                        frontera.append(a)
        return 'Insatisfacible', None

    def ver(self, D):
        vis = []
        A = str(self)
        for c in A:
            if c == '-':
                vis.append(' no ')
            elif c in ['(', ')']:
                vis.append(c)
            elif c in ['>', 'Y', 'O']:
                vis.append(' ' + c + ' ')
            elif c == '=':
                vis.append(' sii ')
            else:
                try:
                    vis.append(D.escribir(c))
                except:
                    raise Exception("¡Caracter inválido!")
        return ''.join(vis)

    def ver_comb(self, lista_descriptores):
        vis = []
        A = str(self)
        contador = 0
        while contador < len(A):
            c = A[contador]
            if c == '-':
                if A[contador + 1] == '(':
                    vis.append(' no ')
                else:
                    try:
                        D = encuentra_descriptor(A[contador + 1], lista_descriptores)
                        vis.append(D.escribir(A[contador:contador + 2]))
                        contador += 1
                    except:
                        raise Exception("¡Caracter inválido!")
            elif c in ['(', ')']:
                vis.append(c)
            elif c in ['>', 'Y', 'O']:
                vis.append(' ' + c + ' ')
            elif c == '=':
                vis.append(' sii ')
            else:
                try:
                    D = encuentra_descriptor(c, lista_descriptores)
                    vis.append(D.escribir(c))
                except:
                    raise Exception("¡Caracter inválido!")
            contador += 1
        return ''.join(vis)

    def eliminar_imp(self):
        if type(self) == Letra:
            return self
        elif type(self) == Negacion:
            return Negacion(self.subf.eliminar_imp())
        elif type(self) == Binario:
            if self.conectivo == '>':
                return Binario('O',
                               Negacion(self.left.eliminar_imp()),
                               self.right.eliminar_imp())
            else:
                return Binario(self.conectivo,
                               self.left.eliminar_imp(),
                               self.right.eliminar_imp())

    def eliminar_doble_imp(self):
        if type(self) == Letra:
            return self
        elif type(self) == Negacion:
            return Negacion(self.subf.eliminar_doble_imp())
        elif type(self) == Binario:
            if self.conectivo == '=':
                return Binario('Y',
                               Binario('O',
                                       Negacion(self.left.eliminar_doble_imp()),
                                       self.right.eliminar_doble_imp()),
                               Binario('O',
                                       Negacion(self.right.eliminar_doble_imp()),
                                       self.left.eliminar_doble_imp()))
            else:
                return Binario(self.conectivo,
                               self.left.eliminar_doble_imp(),
                               self.right.eliminar_doble_imp())

    def eliminar_doble_negacion(self):
        if type(self) == Letra:
            return self
        elif type(self) == Negacion:
            if type(self.subf) == Negacion:
                return deepcopy(self.subf.subf.eliminar_doble_negacion())
            else:
                return Negacion(self.subf.eliminar_doble_negacion())
        elif type(self) == Binario:
            return Binario(self.conectivo,
                           self.left.eliminar_doble_negacion(),
                           self.right.eliminar_doble_negacion())

    def cambiar_de_morgan_y(self):
        if type(self) == Letra:
            return self
        elif type(self) == Negacion:
            if type(self.subf) == Binario:
                if self.subf.conectivo == 'Y':
                    return Binario('O',
                                   Negacion(self.subf.left.cambiar_de_morgan_y()),
                                   Negacion(self.subf.right.cambiar_de_morgan_y()))
                else:
                    return Negacion(self.subf.cambiar_de_morgan_y())
            else:
                return Negacion(self.subf.cambiar_de_morgan_y())
        elif type(self) == Binario:
            return Binario(self.conectivo,
                           self.left.cambiar_de_morgan_y(),
                           self.right.cambiar_de_morgan_y())

    def cambiar_de_morgan_o(self):
        if type(self) == Letra:
            return self
        elif type(self) == Negacion:
            if type(self.subf) == Binario:
                if self.subf.conectivo == 'O':
                    return Binario('Y',
                                   Negacion(self.subf.left.cambiar_de_morgan_o()),
                                   Negacion(self.subf.right.cambiar_de_morgan_o()))
                else:
                    return Negacion(self.subf.cambiar_de_morgan_o())
            else:
                return Negacion(self.subf.cambiar_de_morgan_o())
        elif type(self) == Binario:
            return Binario(self.conectivo,
                           self.left.cambiar_de_morgan_o(),
                           self.right.cambiar_de_morgan_o())

    def distribuir_o_en_y(self):
        if type(self) == Letra:
            return self
        elif type(self) == Negacion:
            return Negacion(self.subf.distribuir_o_en_y())
        elif type(self) == Binario:
            if self.conectivo == 'O':
                if type(self.right) == Binario:
                    if self.right.conectivo == 'Y':
                        B = self.left.distribuir_o_en_y()
                        C = self.right.left.distribuir_o_en_y()
                        D = self.right.right.distribuir_o_en_y()
                        return Binario('Y', Binario('O', B, C), Binario('O', B, D))
                if type(self.left) == Binario:
                    if self.left.conectivo == 'Y':
                        B = self.left.left.distribuir_o_en_y()
                        C = self.left.right.distribuir_o_en_y()
                        D = self.right.distribuir_o_en_y()
                        return Binario('Y', Binario('O', B, D), Binario('O', C, D))
        return Binario(self.conectivo,
                       self.left.distribuir_o_en_y(),
                       self.right.distribuir_o_en_y())

    def fnc(self):
        A = self.eliminar_doble_imp()
        A = A.eliminar_imp()
        A = A.eliminar_doble_negacion()
        A = A.cambiar_de_morgan_y()
        A = A.cambiar_de_morgan_o()
        A = A.eliminar_doble_negacion()
        aux = A.distribuir_o_en_y()
        while str(A) != str(aux):
            A = deepcopy(aux)
            aux = A.distribuir_o_en_y()
        return aux


class Letra(Formula):
    def __init__(self, letra: str):
        self.letra = letra


class Negacion(Formula):
    def __init__(self, subf: Formula):
        self.subf = subf


class Binario(Formula):
    def __init__(self, conectivo: str, left: Formula, right: Formula):
        assert conectivo in ['Y', 'O', '>', '=']
        self.conectivo = conectivo
        self.left = left
        self.right = right


# 1.2  Convertr

def inorder_to_tree(cadena: str):
    conectivos = ['Y', 'O', '>', '=']
    if len(cadena) == 1:
        return Letra(cadena)
    elif cadena[0] == '-':
        return Negacion(inorder_to_tree(cadena[1:]))
    elif cadena[0] == '(':
        counter = 0
        for i in range(1, len(cadena)):
            if cadena[i] == '(':
                counter += 1
            elif cadena[i] == ')':
                counter -= 1
            elif cadena[i] in conectivos and counter == 0:
                return Binario(cadena[i],
                               inorder_to_tree(cadena[1:i]),
                               inorder_to_tree(cadena[i + 1:-1]))
    else:
        raise Exception('¡Cadena inválida!')


def polaca_inversa_to_tree(A, letrasProposicionales):
    conectivos = ['O', 'Y', '>', '=']
    pila = []
    for c in A:
        if c in letrasProposicionales:
            pila.append(Letra(c))
        elif c == '-':
            aux = Negacion(pila[-1])
            del pila[-1]
            pila.append(aux)
        elif c in conectivos:
            aux = Binario(c, pila[-1], pila[-2])
            del pila[-1]
            del pila[-1]
            pila.append(aux)
    return pila[-1]



# 1.3  Descriptor


class Descriptor:
    """
    Codifica un descriptor de N argumentos en un solo carácter Unicode.
    Métodos originales conservados: P() e inv().
    Se añaden aliases ravel()/unravel() para compatibilidad con reglas.py.
    """

    def __init__(self, args_lista, chrInit=256):
        self.args_lista = args_lista
        assert len(args_lista) > 0, "Debe haber por lo menos un argumento"
        self.chrInit = chrInit
        self.rango = [chrInit, chrInit + int(np.prod(self.args_lista))]

    def check_lista_valores(self, lista_valores):
        for i, v in enumerate(lista_valores):
            assert v >= 0, "Valores deben ser no negativos"
            assert v < self.args_lista[i], \
                f"Valor debe ser menor a {self.args_lista[i]}"

    def codifica(self, lista_valores):
        self.check_lista_valores(lista_valores)
        cod = lista_valores[0]
        n_columnas = 1
        for i in range(len(lista_valores) - 1):
            n_columnas *= self.args_lista[i]
            cod = n_columnas * lista_valores[i + 1] + cod
        return cod

    def decodifica(self, n):
        decods = []
        if len(self.args_lista) > 1:
            for i in range(len(self.args_lista) - 1):
                n_columnas = int(np.prod(self.args_lista[:-(i + 1)]))
                decods.insert(0, int(n / n_columnas))
                n = n % n_columnas
        decods.insert(0, n % self.args_lista[0])
        return decods

    # Nombres originales de Logica_5.py
    def P(self, lista_valores):
        return chr(self.chrInit + self.codifica(lista_valores))

    def inv(self, codigo):
        return self.decodifica(ord(codigo) - self.chrInit)

    # Aliases usados en reglas.py (y aquí en Euler)
    def ravel(self, lista_valores):
        return self.P(lista_valores)

    def unravel(self, codigo):
        return self.inv(codigo)


def encuentra_descriptor(c, lista_descriptores):
    for D in lista_descriptores:
        if D.rango[0] <= ord(c) < D.rango[1]:
            return D
    raise Exception(f"No se encontró descriptor para el carácter {c}")



# 1.4  Y-toria / O-toria



def Ytoria(lista_forms):
    form = ''
    inicial = True
    for f in lista_forms:
        if inicial:
            form = f
            inicial = False
        else:
            form = '(' + form + 'Y' + f + ')'
    return form


def Otoria(lista_forms):
    form = ''
    inicial = True
    for f in lista_forms:
        if inicial:
            form = f
            inicial = False
        else:
            form = '(' + form + 'O' + f + ')'
    return form



# 1.5  Tableaux


class nodos_tableaux:

    def __init__(self, fs):
        clasfs = [(A, str(A), *A.clasifica_para_tableaux()) for A in fs]
        self.alfas     = [c for c in clasfs if c[3] == 'alfa']
        self.betas     = [c for c in clasfs if c[3] == 'beta']
        self.literales = [c for c in clasfs if c[3] == 'literal']

    def __str__(self):
        cadena  = f'Alfas:{[str(c[1]) for c in self.alfas]}\n'
        cadena += f'Betas:{[str(c[1]) for c in self.betas]}\n'
        cadena += f'Literales:{[str(c[1]) for c in self.literales]}'
        return cadena

    def tiene_lit_comp(self):
        lits   = [c[1] for c in self.literales]
        l_pos  = [l for l in lits if '-' not in l]
        l_negs = [l[1:] for l in lits if '-' in l]
        return len(set(l_pos).intersection(set(l_negs))) > 0

    def es_hoja(self):
        if self.tiene_lit_comp():
            return 'cerrada'
        elif len(self.alfas) == 0 and len(self.betas) == 0:
            return 'abierta'
        return None

    def interp(self):
        I = {}
        for lit in self.literales:
            l = lit[1]
            if '-' not in l:
                I[l] = True
            else:
                I[l[1:]] = False
        return I

    def expandir(self):
        f_alfas     = deepcopy(self.alfas)
        f_betas     = deepcopy(self.betas)
        f_literales = deepcopy(self.literales)

        if len(self.alfas) > 0:
            f, s, num_regla, cl = f_alfas.pop(0)
            if num_regla == 1:
                formulas = [f.subf.subf]
            elif num_regla == 2:
                formulas = [f.left, f.right]
            elif num_regla == 3:
                formulas = [Negacion(f.subf.left), Negacion(f.subf.right)]
            elif num_regla == 4:
                formulas = [f.subf.left, Negacion(f.subf.right)]
            for nueva_f in formulas:
                clasf = nueva_f.clasifica_para_tableaux()
                lista = f_alfas if clasf[1] == 'alfa' else \
                        f_betas if clasf[1] == 'beta' else f_literales
                if str(nueva_f) not in [c[1] for c in lista]:
                    lista.append((nueva_f, str(nueva_f), *clasf))
            nuevo = nodos_tableaux([])
            nuevo.alfas     = f_alfas
            nuevo.betas     = f_betas
            nuevo.literales = f_literales
            return [nuevo, None]

        elif len(self.betas) > 0:
            f, s, num_regla, cl = f_betas.pop(0)
            if num_regla == 1:
                B1, B2 = Negacion(f.subf.left), Negacion(f.subf.right)
            elif num_regla == 2:
                B1, B2 = f.left, f.right
            elif num_regla == 3:
                B1, B2 = Negacion(f.left), f.right

            f_alfas2     = deepcopy(f_alfas)
            f_betas2     = deepcopy(f_betas)
            f_literales2 = deepcopy(f_literales)

            for B, fa, fb, fl in [(B1, f_alfas,  f_betas,  f_literales),
                                   (B2, f_alfas2, f_betas2, f_literales2)]:
                clasf = B.clasifica_para_tableaux()
                lista = fa if clasf[1] == 'alfa' else \
                        fb if clasf[1] == 'beta' else fl
                if str(B) not in [c[1] for c in lista]:
                    lista.append((B, str(B), *clasf))

            n1 = nodos_tableaux([]); n1.alfas = f_alfas;  n1.betas = f_betas;  n1.literales = f_literales
            n2 = nodos_tableaux([]); n2.alfas = f_alfas2; n2.betas = f_betas2; n2.literales = f_literales2
            return [n1, n2]

        return [None, None]



# 1.6  Tseitin


def a_clausal(A):
    """
    Subrutina de Tseitin: convierte una subfórmula atómica de la pila
    (p=-q | p=(qYr) | p=(qOr) | p=(q>r) | p=(q=r)) a forma clausal.
    """
    assert len(A) == 4 or len(A) == 7, "Fórmula incorrecta"
    p = A[0]
    if '-' in A:
        q = A[-1]
        B = '-' + p + 'O-' + q + 'Y' + p + 'O' + q
    elif 'Y' in A:
        q, r = A[3], A[5]
        B = q + 'O-' + p + 'Y' + r + 'O-' + p + 'Y-' + q + 'O-' + r + 'O' + p
    elif 'O' in A:
        q, r = A[3], A[5]
        B = '-' + q + 'O' + p + 'Y-' + r + 'O' + p + 'Y' + q + 'O' + r + 'O-' + p
    elif '>' in A:
        q, r = A[3], A[5]
        B = q + 'O' + p + 'Y-' + r + 'O' + p + 'Y-' + q + 'O' + r + 'O-' + p
    elif '=' in A:
        q, r = A[3], A[5]
        B = (q + 'O-' + r + 'O-' + p + 'Y-' + q + 'O' + r + 'O-' + p +
             'Y-' + q + 'O-' + r + 'O' + p + 'Y' + q + 'O' + r + 'O' + p)
    else:
        raise Exception('Error en a_clausal(): Fórmula incorrecta')
    return [c.split('O') for c in B.split('Y')]


def tseitin(A):
    """
    Transformación de Tseitin.
    Input : A (str) fórmula en notación inorder.
    Output: lista de cláusulas (list[list[str]]).
    """
    f = inorder_to_tree(A)
    letrasp = f.letras()
    cods_letras = [ord(x) for x in letrasp]
    m = max(cods_letras)
    letrasp_tseitin = [chr(x) for x in range(m, m + f.num_conec())]
    letrasp = list(letrasp) + letrasp_tseitin

    L, Pila, i = [], [], -1
    s = A[0]
    while len(A) > 0:
        if (s in letrasp) and (len(Pila) > 0) and (Pila[-1] == '-'):
            i += 1
            atomo = letrasp_tseitin[i]
            Pila = Pila[:-1]
            Pila.append(atomo)
            L.append(atomo + '=-' + s)
            A = A[1:]
            if len(A) > 0:
                s = A[0]
        elif s == ')':
            w, O, v = Pila[-1], Pila[-2], Pila[-3]
            Pila = Pila[:len(Pila) - 4]
            i += 1
            atomo = letrasp_tseitin[i]
            L.append(atomo + '=(' + v + O + w + ')')
            s = atomo
        else:
            Pila.append(s)
            A = A[1:]
            if len(A) > 0:
                s = A[0]

    atomo = Pila[-1] if i < 0 else letrasp_tseitin[i]
    B = [[[atomo]]] + [a_clausal(x) for x in L]
    return [val for sublist in B for val in sublist]



# 1.7  Utilidades SAT


def interpretacion_aleatoria(letrasp):
    return {p: randint(0, 1) == 1 for p in letrasp}


def complemento(l):
    """Devuelve el complemento de un literal (str)."""
    return l[1:] if '-' in l else '-' + l


def flip_literal(I, l):
    """
    Invierte el valor del átomo de l en la interpretación I.
    Nota: asume que el átomo es el último carácter de l (válido mientras
    los átomos sean caracteres Unicode de longitud 1 del Descriptor).
    """
    p = l[-1]
    Ip = deepcopy(I)
    Ip[p] = not I[p]
    return Ip


def eliminar_literal(S, l):
    S1 = [c for c in S if l not in c]
    lc = complemento(l)
    return [[p for p in c if p != lc] for c in S1]


def extender_I(I, l):
    I1 = {k: I[k] for k in I if k != l}
    if '-' in l:
        I1[l[1:]] = False
    else:
        I1[l] = True
    return I1



# 1.8  Unit Propagation + DPLL


def unit_propagate(S, I):
    """
    Propaga cláusulas unitarias hasta que no queden o S sea insatisfacible.
    Input / Output: S (list[list[str]]), I (dict átomo->bool).
    """
    while [] not in S:
        l = ''
        for x in S:
            if len(x) == 1:
                l = x[0]
                S = eliminar_literal(S, l)
                I = extender_I(I, l)
                break
        if l == '':
            break
    return S, I


def dpll(S, I):
    """
    DPLL recursivo con unit propagation.
    Input : S (list[list[str]]), I (dict átomo->bool).
    Output: ('Satisfacible'|'Insatisfacible', I).
    """
    S, I = unit_propagate(S, I)
    if len(S) == 0:
        return 'Satisfacible', I
    if [] in S:
        return 'Insatisfacible', {}
    l  = choice(choice(S))
    lc = complemento(l)
    sat, newI = dpll(eliminar_literal(S, l),  extender_I(I, l))
    if sat == 'Satisfacible':
        return sat, newI
    return dpll(eliminar_literal(S, lc), extender_I(I, lc))



# 1.9  WalkSAT


class WalkSatEstado:
    """
    Estado interno de WalkSAT.
    Mantiene la interpretación actual y las cláusulas sat/unsat.
    """

    def __init__(self, S):
        self.S = S
        self.letrasp = list(set(l[-1] for C in self.S for l in C))
        self.I = interpretacion_aleatoria(self.letrasp)
        self._recalcular()

    def _recalcular(self):
        self.I_lits = set(
            [p for p in self.letrasp if self.I[p]] +
            ['-' + p for p in self.letrasp if not self.I[p]]
        )
        self.clausulas_sat   = [C for C in self.S if any(x in self.I_lits for x in C)]
        self.clausulas_unsat = [C for C in self.S if C not in self.clausulas_sat]

    def actualizar(self, I):
        self.I = I
        self._recalcular()

    def SAT(self):
        return len(self.clausulas_unsat) == 0

    def break_count(self, l):
        """
        Número de cláusulas satisfechas que quedarían insatisfechas si
        se hace flip de l.
        """
        lit = l if l in self.I_lits else complemento(l)
        return sum(
            1 for C in self.clausulas_sat
            if set(C).intersection(self.I_lits) == {lit}
        )


def walkSAT(S, max_flips=1000, max_tries=100, p=0.5):
    """
    WalkSAT estándar.
    Input : S (list[list[str]]) conjunto de cláusulas.
    Output: ('Satisfacible', I) o (None, {}).
    """
    w = WalkSatEstado(S)
    for _ in range(max_tries):
        w.actualizar(interpretacion_aleatoria(w.letrasp))
        for _ in range(max_flips):
            if w.SAT():
                return 'Satisfacible', w.I
            C      = choice(w.clausulas_unsat)
            breaks = sorted([(l, w.break_count(l)) for l in C], key=lambda x: x[1])
            min_bc = breaks[0]
            if min_bc[1] == 0:
                v = min_bc[0]
            elif uniform(0, 1) < p:
                v = choice(C)
            else:
                v = min_bc[0]
            w.actualizar(flip_literal(w.I, v))
    return None, {}



# 2. CLASE EULER — reglas en formato clausal optimizado


def _escribir_euler(descriptor, literal):
    """Función de visualización para el descriptor Va."""
    if '-' in literal:
        atomo, neg = literal[1:], 'NO '
    else:
        atomo, neg = literal, ''
    nodos = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
    x, y, t = descriptor.unravel(atomo)
    return f"{neg}Va({nodos[x]},{nodos[y]},t={t+1})"


class Euler:
    """
    Modelo SAT del sendero Euleriano sobre el grafo del enunciado.
    Todas las reglas producen directamente listas de cláusulas
    (list[list[str]]) listas para DPLL y WalkSAT.
    """

    def __init__(self):
        self.vertices = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
        self.idx      = {v: i for i, v in enumerate(self.vertices)}

        # Aristas no dirigidas del grafo del enunciado
        self.edges = [
            ('A', 'B'), ('A', 'E'), ('B', 'C'), ('B', 'F'),
            ('C', 'D'), ('C', 'G'), ('C', 'F'),
            ('D', 'H'), ('D', 'G'),
            ('E', 'F'),
            ('F', 'G'),
            ('G', 'H'),
        ]

        # Aristas dirigidas (ambas orientaciones)
        self.D = [(i, j) for e in self.edges for i, j in [e, (e[1], e[0])]]

        # Descriptor Va(nodo_i, nodo_j, turno)  — índices base-0
        self.Va = Descriptor([8, 8, 12])
        self.Va.escribir = MethodType(_escribir_euler, self.Va)

        # Cláusulas totales
        self.clausulas = (
            self.R1() + self.R2() + self.R3() +
            self.R4() + self.R5() +
            self.R6_existencia() + self.R6_unicidad()
        )


    #  Helper interno                                                      


    def _va(self, i, j, t):
        """
        Literal positivo Va(i, j, t).
        i, j: nombres de vértice ('A'..'H').  t: turno lógico 1-12.
        Internamente el turno se indexa 0-11.
        """
        return self.Va.ravel([self.idx[i], self.idx[j], t - 1])

    def _neg(self, literal):
        return '-' + literal


    #  R1 — El recorrido empieza en H (turno 1)                          
    #                                                                      
    #  OR_{(H,j)∈D}  Va(H,j,1)                                           
    #  → una sola cláusula disyuntiva                                     


    def R1(self):
        return [[self._va('H', j, 1) for (i, j) in self.D if i == 'H']]


    #  R2 — El recorrido termina en B (turno 12)                         
    #                                                                      
    #  OR_{(i,B)∈D}  Va(i,B,12)                                          
    #  → una sola cláusula disyuntiva                                     


    def R2(self):
        return [[self._va(i, 'B', 12) for (i, j) in self.D if j == 'B']]


    #  R3 — Continuidad                                                   
    #                                                                      
    #  AND_{t=1}^{11} AND_{(i,j)∈D}                                      
    #      Va(i,j,t) → OR_{(j,k)∈D} Va(j,k,t+1)                         
    #                                                                      
    #  Reescritura clausal directa:                                        
    #      A → (B1 O B2 O ... O Bn)  ≡  [-A, B1, B2, ..., Bn]           
    #  → una cláusula por cada par (arista dirigida, turno)               
    
    def R3(self):
        clausulas = []
        for t in range(1, 12):
            for (i, j) in self.D:
                ant_neg     = self._neg(self._va(i, j, t))
                consecuentes = [self._va(j, k, t + 1) for (j2, k) in self.D if j2 == j]
                clausulas.append([ant_neg] + consecuentes)
        return clausulas


    #  R4 — Cada arista se cruza al menos una vez                        
    #                                                                      
    #  AND_{{i,j}∈P}  OR_{t=1}^{12}  (Va(i,j,t) O Va(j,i,t))           
    #  → cláusula plana por arista no dirigida                            


    def R4(self):
        clausulas = []
        for (i, j) in self.edges:
            clausula = []
            for t in range(1, 13):
                clausula.append(self._va(i, j, t))
                clausula.append(self._va(j, i, t))
            clausulas.append(clausula)
        return clausulas

    
    #  R5 — Ninguna arista se cruza más de una vez                     
    #                                                                    
    #  ¬((Va(i,j,t) O Va(j,i,t)) ∧ (Va(i,j,s) O Va(j,i,s)))          
    #                                                                    
    #  De Morgan + distribución:                                         
    #  (-a O -c) ∧ (-a O -d) ∧ (-b O -c) ∧ (-b O -d)                 
    #  → 4 cláusulas binarias por cada par de turnos (t,s)             
    

    def R5(self):
        clausulas = []
        for (i, j) in self.edges:
            for t, s in combinations(range(1, 13), 2):
                a = self._va(i, j, t); b = self._va(j, i, t)
                c = self._va(i, j, s); d = self._va(j, i, s)
                clausulas.append([self._neg(a), self._neg(c)])
                clausulas.append([self._neg(a), self._neg(d)])
                clausulas.append([self._neg(b), self._neg(c)])
                clausulas.append([self._neg(b), self._neg(d)])
        return clausulas

    
    #  R6 Existencia — en cada turno ocurre al menos un recorrido      
    #                                                                    
    #  AND_{t=1}^{12}  OR_{(i,j)∈D}  Va(i,j,t)                        
    #  → una cláusula disyuntiva por turno                              
    

    def R6_existencia(self):
        return [[self._va(i, j, t) for (i, j) in self.D] for t in range(1, 13)]

    
    #  R6 Unicidad — en cada turno ocurre exactamente un recorrido     
    #                                                                    
    #  Va(i,j,t) → ¬Va(k,l,t)   para todo (k,l)≠(i,j)                
    #  ≡  [-Va(i,j,t), -Va(k,l,t)]   (cláusula binaria at-most-one)   
    #  → C(|D|,2) × 12 cláusulas de tamaño 2                          
    

    def R6_unicidad(self):
        clausulas = []
        for t in range(1, 13):
            for idx_ij, (i, j) in enumerate(self.D):
                for idx_kl, (k, l) in enumerate(self.D):
                    if idx_kl > idx_ij:
                        clausulas.append([self._neg(self._va(i, j, t)),
                                          self._neg(self._va(k, l, t))])
        return clausulas


    #  Información de diagnóstico                                     

    def info(self):
        partes = [
            ('R1  inicio en H',      self.R1()),
            ('R2  fin en B',         self.R2()),
            ('R3  continuidad',      self.R3()),
            ('R4  al menos 1 vez',   self.R4()),
            ('R5  a lo sumo 1 vez',  self.R5()),
            ('R6  existencia',       self.R6_existencia()),
            ('R6  unicidad',         self.R6_unicidad()),
        ]
        print(" " * 50)
        print("  MODELO CLAUSAL — SENDERO EULERIANO")
        print(" " * 50)
        total = 0
        for nombre, cs in partes:
            print(f"  {nombre:<22} {len(cs):>5} cláusula(s)")
            total += len(cs)
        print(" " * 50)
        print(f"  TOTAL                  {total:>5} cláusulas")
        atomos = len(set(l[-1] for c in self.clausulas for l in c))
        print(f"  Átomos distintos       {atomos:>5}")
        print(" " * 50)

# 3. WALKSAT SOBRE R1, R2, R3 


def walksat_regla(nombre, clausulas, euler, max_flips=500, max_tries=50, p=0.5):
    """
    Aplica WalkSAT a un subconjunto de cláusulas e imprime el resultado.
    """
    print(f"  WalkSAT sobre {nombre}")
    print(f"  Cláusulas: {len(clausulas)}      "
          f"Átomos: {len(set(l[-1] for c in clausulas for l in c))}")

    resultado, I = walkSAT(clausulas, max_flips=max_flips,
                           max_tries=max_tries, p=p)

    if resultado == 'Satisfacible':
        print(f"  Resultado : Satisfacible")
        verdaderos = []
        for atomo, valor in sorted(I.items(), key=lambda x: euler.Va.unravel(x[0])[2]):
            if valor:
                x, y, t = euler.Va.unravel(atomo)
                nodos = euler.vertices
                verdaderos.append(f"Va({nodos[x]},{nodos[y]},t={t+1})")
        for v in verdaderos:
            print(f"    {v}")
    else:
        print("  Resultado : No encontró solución en los límites dados.")
        print("  Intente aumentar max_flips o max_tries.")

    return resultado, I



# 4. MAIN


if __name__ == '__main__':

    euler = Euler()
    euler.info()

   
    # WalkSAT sobre R1 sola
    # "¿Existe algún recorrido que parta de H en el turno 1?"
  
    walksat_regla("R1  (inicio en H, turno 1)",
                  euler.R1(),
                  euler,
                  max_flips=200, max_tries=20)

    
    # WalkSAT sobre R2 sola
    # "¿Existe algún recorrido que llegue a B en el turno 12?"
   
    walksat_regla("R2  (fin en B, turno 12)",
                  euler.R2(),
                  euler,
                  max_flips=200, max_tries=20)

    
    # WalkSAT sobre R3 sola
    # "¿Existe una secuencia de 11 movimientos consecutivos válidos?"
    # Nota: sin R6-unicidad, WalkSAT puede satisfacer R3 repitiendo aristas.
    # Eso es correcto: R3 solo garantiza continuidad, no inyectividad.
    
    walksat_regla("R3  (continuidad, t=1..11)",
                  euler.R3(),
                  euler,
                  max_flips=2000, max_tries=100)

    
    # WalkSAT sobre R1 + R2 + R3 combinadas
    # "¿Existe un camino continuo de H a B en 12 pasos?"
    
    walksat_regla("R1 + R2 + R3  (camino H→B continuo)",
                  euler.R1() + euler.R2() + euler.R3(),
                  euler,
                  max_flips=5000, max_tries=200)
```
